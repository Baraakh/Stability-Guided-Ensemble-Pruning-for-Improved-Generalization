# Results Display

Formatted tables from the ensemble pruning pipeline.  
`*` = best (max) value in column | `-` = worst (min) value in column  
For `Std` columns, lower is better: `*` = min, `-` = max.

```python
import pandas as pd
from IPython.display import display, HTML

def style_table(df, title, index_col=None):
    """Display a DataFrame with * for best and - for worst per numeric column."""
    if index_col and index_col in df.columns:
        df = df.set_index(index_col)

    numeric_cols = df.select_dtypes(include='number').columns

    def fmt_cell(val, col, best_val, worst_val):
        s = f'{val:.4f}'
        if val == best_val:
            return s + ' *'
        if val == worst_val:
            return s + ' -'
        return s

    display_df = df.copy()
    for col in numeric_cols:
        lower_is_better = 'Std' in col
        best = df[col].min() if lower_is_better else df[col].max()
        worst = df[col].max() if lower_is_better else df[col].min()
        display_df[col] = df[col].apply(lambda v: fmt_cell(v, col, best, worst))

    def highlight(val):
        if isinstance(val, str):
            if val.endswith(' *'):
                return 'background-color: #d4edda; font-weight: bold'
            if val.endswith(' -'):
                return 'background-color: #f8d7da'
        return ''

    styled = (display_df.style
              .set_caption(title)
              .map(highlight)
              .set_table_styles([{
                  'selector': 'caption',
                  'props': [('font-size', '16px'), ('font-weight', 'bold'), ('padding', '8px')]
              }]))
    display(styled)

print('Ready.')
```

## Stability Results

```python
style_table(
    pd.read_csv('datasets/stability_results_heart.csv'),
    'Stability Results — Heart Disease',
    index_col='Model'
)
```

```python
style_table(
    pd.read_csv('datasets/stability_results_pima.csv'),
    'Stability Results — Pima Diabetes',
    index_col='Model'
)
```

## Final Results

```python
style_table(
    pd.read_csv('datasets/final_results_heart.csv'),
    'Final Results — Heart Disease',
    index_col='Method'
)
```
